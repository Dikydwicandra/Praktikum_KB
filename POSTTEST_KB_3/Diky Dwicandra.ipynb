#### Library

```python
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
```

#### Membaca data

```python
df=pd.read_csv("HospInfo.csv")
```

```python
df.head()
```

#### Bentuk Grafik Batang

```python
sns.countplot(data=df,x="Hospital Type",palette="bright")
plt.xlabel("Tipe Rumah Sakit")
plt.ylabel("Jumlah Rumah Sakit")
plt.title("Penyebaran Jumlah Tipe Rumah Sakit")
plt.show()
```

#### KESIMPULAN:
##### kesimpulannya adalah tipe rumah sakit Acute Care Hospitals memiliki jumlah terbanyak diatas 3000, dan tipe rumah sakit Childrens memiliki jumlah terkecil yaitu dibawah 500.

## Grafik Pie

```python
data_pie = df.loc[0:300, "Hospital Type"].value_counts().rename_axis("Hospital Type").reset_index(name="Jumlah_Data")
data_pie.head()
```

```python
plt.pie(
    data_pie["Jumlah_Data"],
    labels = data_pie["Hospital Type"],
    autopct = '%.2f%%')
plt.title("Perbandingan Jumlah Tipe Rumah Sakit", fontsize = 10)
plt.show()
```

## Kesimpulan
#### Perbandingan jumlah rumah sakit pada 300 data pertama, Acute Care Hospitals 63.79%, dan Childrens 2.99%, dan untuk Critical Access Hospitals 33.22%

## Grafik Boxplot

```python
plt.figure(figsize=(10,10))
sns.boxplot(data = df, x = "Hospital Type" ,y = "ZIP Code")
plt.xlabel("Tipe Rumah Sakit")
plt.ylabel("ZIP Code")
plt.title("Penyebaran ZIP Code Tipe Rumah Sakit")
plt.show()
```

## Kesimpulan
#### Boxplot terbentuk dengan nilai: 1. min dibawah 40000 dan max diatas 60000 dengan median diantara 40000 dan 60000 pada Acute Care Hospital, 2. min dibawah 60000 dan max diatas 60000 dengan median di 60000 pada Critical Access Hospitals, 3. min dibawah 40000 dan max di 80000 dengan median diantara 60000 dan 80000 pada Childrens Hospitals.
