### import library yang akan digunakan

### Mengakses file csv

```python
datasplit = pd.read_csv("applemobilitytrendsRusak.csv")
datasplit.head()
```

### Melakukan Spliting menjadi 2

```python
x = datasplit.iloc[:, :-1]
y = datasplit.iloc[:, -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```
